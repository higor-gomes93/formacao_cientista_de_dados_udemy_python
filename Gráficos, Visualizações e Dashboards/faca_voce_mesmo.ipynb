```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
# Carregamento da base de dados
base = pd.read_csv('dados.csv', sep = ";")
base.shape
```

```python
base.head()
```

```python
# Gráfico de distribuição para o PIB
srn.distplot(base.PIB, bins = 10, axlabel = 'PIB').set_title('PIB')
```

```python
# Gráfico de distribuição para o valor do empenho
srn.distplot(base.VALOREMPENHO, bins = 10, axlabel = 'Valor do Empenho').set_title('Valor do Empenho')
```

```python
# Gráfico de barras dos 10 maiores PIBs por município
agrupado_1 = base.sort_values('PIB').head(10)
agrupado_1 = agrupado_1.iloc[:, 1:3] # Pega as colunas 1 e 2, para no 3
agrupado_1
agrupado_1.plot.bar(x = 'MUNICIPIO', y = 'PIB', color = 'grey')
```

```python
# Gráfico de barras dos 10 maiores valores de empenho por município
agrupado_1 = base.sort_values('VALOREMPENHO').head(10)
agrupado_1 = agrupado_1.iloc[:, [1, 3]] # Pega as colunas 1 e 2, para no 3
agrupado_1
agrupado_1.plot.bar(x = 'MUNICIPIO', y = 'VALOREMPENHO', color = 'grey')
```
